## Data Projections


### PCA FOR QUALIFICATIONS ACCORDING TO OCCUPATIONS

```python
pivot_df = westminster_df[['Qualification_Code','Qualification_Code_Description','Occupation_Code','Occupation_Code_Description','Observation']]
```

```python
pivot_df = pivot_df.groupby(by=['Qualification_Code','Occupation_Code_Description'])['Observation'].sum().reset_index()
```

```python
pivot_df = pivot_df.pivot(index='Qualification_Code',columns='Occupation_Code_Description',values='Observation')
```

```python
pivot_df_copy = pivot_df.copy()
```

```python
pivot_df.reset_index(drop=True,inplace=True)
```

#### Perform Standard scaling of the pivot df

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
pivot_df_scaled = scaler.fit_transform(pivot_df)
```

```python
pivot_df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(pivot_df_scaled)
```

```python
# Creating a DataFrame for the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Qualification_Code'] = westminster_df['Qualification_Code'].unique()
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'], alpha=0.8)
plt.title('PCA of Occupation by Qualification_Code')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pca_df['Qualification_Code']):
    plt.annotate(txt, (pca_df['Principal Component 1'][i], pca_df['Principal Component 2'][i]), fontsize=8)
plt.grid(True)
```

```python
# pca_df.to_csv('./Results/pca_df_occupation_by_qualification.csv',index=False)
```

```python
# K-means on pca_df
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X = pca_df[['Principal Component 1', 'Principal Component 2']]

num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

kmeans_labels = kmeans.fit_predict(X)

pca_df['Cluster'] = kmeans_labels

# Visualize the clusters
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X['Principal Component 1'], X['Principal Component 2'], c=kmeans_labels, cmap='viridis', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clusters from PCA Results (KMeans)')
for i, txt in enumerate(pca_df['Qualification_Code']):
    plt.annotate(txt, (pca_df['Principal Component 1'][i], pca_df['Principal Component 2'][i]), fontsize=8)
plt.grid(True)
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()
```

```python
X = pca_df[['Principal Component 1', 'Principal Component 2']]
inertias = []
K = range(1, 6) 

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plotting the results
plt.figure(figsize=(8, 4))
plt.plot(K, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.title('Elbow Method showing the optimal k')
plt.show()
```

### PCA FOR OCCUPATION ACCORDING TO QUALIFICATION

```python
pivot_df2 = westminster_df[['Qualification_Code','Qualification_Code_Description','Occupation_Code','Occupation_Code_Description','Observation']]
```

```python
pivot_df2 = pivot_df2.groupby(by=['Qualification_Code_Description','Occupation_Code'])['Observation'].sum().reset_index()
```

```python
pivot_df2 = pivot_df2.pivot(index='Occupation_Code',columns='Qualification_Code_Description',values='Observation')
```

```python
pivot_df2_copy = pivot_df2.copy()
```

```python
pivot_df2.reset_index(drop=True,inplace=True)
```

#### Perform Standard scaling of the pivot df

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
pivot_df2_scaled = scaler.fit_transform(pivot_df2)
```

```python
# Performing PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Reduce to 2 dimensions for easy visualization
principal_components = pca.fit_transform(pivot_df2_scaled)
```

```python
# Creating a DataFrame for the principal components
pca_df2 = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df2['Occupation_Code'] = westminster_df['Occupation_Code'].unique()
```

```python
pca_df2['Occupation_Code_Description'] = westminster_df['Occupation_Code_Description'].unique()
```

```python
# Plotting the PCA results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_df2['Principal Component 1'], pca_df2['Principal Component 2'], alpha=0.8)
plt.title('PCA of Occupation by Qualification_Code')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pca_df2['Occupation_Code_Description']):
    plt.annotate(txt, (pca_df2['Principal Component 1'][i], pca_df2['Principal Component 2'][i]), fontsize=8)
plt.grid(True)
```

```python
# pca_df2.to_csv('./Results/pca_df_occupation_by_qualification.csv',index=False)
```

#### K-means clustering

#### Using Elbow method to find the best number of clusters to use for the above dataset

```python
X = pca_df2[['Principal Component 1', 'Principal Component 2']]
# Trying different numbers of clusters and recording the inertia
inertias = []
K = range(1, 11)  # Examining 1 through 10 clusters

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plotting the results
plt.figure(figsize=(8, 4))
plt.plot(K, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.title('Elbow Method showing the optimal k')
plt.show()
```

According to the above analysis, 2 is the best number of clusters to be used for clustering

### Using UMAP for the same 

```python
import umap
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
pivot_df3_scaled = scaler.fit_transform(pivot_d3_df)
```

```python
pivot_df3_scaled[:2]
```

```python
for n in (2, 5, 10, 20, 50, 100, 200):
    reducer = umap.UMAP(n_neighbors=n,random_state=42)
    embedding = reducer.fit_transform(pivot_df3_scaled)
    umap_df = pd.DataFrame(data=embedding, columns=['X', 'Y'])
    umap_df.index = pivot_d3_df_copy.index
    umap_df = umap_df.reset_index()
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(umap_df['X'], umap_df['Y'],alpha=0.8)
    plt.title(f'UMAP of Qualification, Age by Occupation, neighbors = {n}')
    plt.xlabel('X')
    plt.ylabel('Y')
    for i, txt in enumerate(umap_df['Age_Code_Description']):
        plt.annotate(txt, (umap_df['X'][i], umap_df['Y'][i]), fontsize=8)
```

```python
reducer = umap.UMAP(n_neighbors=10,random_state=42)
embedding = reducer.fit_transform(pivot_df3_scaled)
umap_df = pd.DataFrame(data=embedding, columns=['X', 'Y'])
umap_df.index = pivot_d3_df_copy.index
umap_df = umap_df.reset_index()
```
